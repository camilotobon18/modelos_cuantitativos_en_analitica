### Función para limpiar caracteres especiales en valores númericos

```python
def limpiar_valor(valor):
    valor = str(valor)
    if valor == np.nan:
        valor = 0
    else:
        valor = valor.replace("$", "")
        valor = valor.replace(",", "")
        valor = float(valor)
    return valor
```

### Función para convertir booleanos a binarios ("t" a 1 y "f" a 0). 

```python
def convertir_binario(valor):
    valor = str(valor)
    valor = valor.lower()
    valor = valor.replace("t", "1")
    valor = valor.replace("f", "0")
    valor = int(valor)
    return valor
```

### Función para restar dos fechas y sacar el número de dias

```python
def num_dias(fecha_inicial, fecha_final):
    fecha_final = pd.to_datetime(fecha_final, dayfirst=True)
    dias = fecha_final - fecha_inicial
    dias = dias.astype(int)
    return dias
```

### Función para contruir un modelo de regresión lineal por medio de statsmodels.api

```python
def build_model(X,y):
    X = sm.add_constant(X) #Adding the constant
    lm = sm.OLS(y,X).fit() # fitting the model
    print(lm.summary()) # model summary
    return X
```

```python
X_train = build_model(X_train,y_train)
```

### Evaluación del modelo

```python
modelNNRegr.predict(X_test)
```

```python
predict_train = modelNNRegr.predict(X_train)
predict_test = modelNNRegr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
print(modelNNRegr.score(X_train,y_train))
print(r2_score(y_test,predict_test))
```

### Evaluación del modelo
El modelo mejora respecto al anterior, ya que el R2 da un valor positivo, sin embargo el valor del RMSE es alto, por lo cual la predicción no es tan exacta.

```python
prediccionesNN = modelNNRegr.predict(X_test)


mse_NN = mean_squared_error (y_true  = y_test, y_pred  = prediccionesNN, squared = True)
rmse_NN = mean_squared_error (y_true  = y_test, y_pred  = prediccionesNN, squared = False)
MAE_NN = mean_absolute_error (y_true  = y_test, y_pred  = prediccionesNN)
MedAE_NN = median_absolute_error (y_true  = y_test, y_pred  = prediccionesNN)
r2_NN = r2_score (y_true  = y_test, y_pred  = prediccionesNN)

print(f"El error MSE   de test es: {mse_NN}")
print(f"El error RMSE  de test es: {rmse_NN}")
print(f"El error MAE   de test es: {MAE_NN}")
print(f"El error MedAE de test es: {MedAE_NN}")
print(f"El error R2    de test es: {r2_NN}")
```
